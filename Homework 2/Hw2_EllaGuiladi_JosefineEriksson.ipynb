# [Solutions]
1- The following equation $\mathcal{L}(\mathbf{w}) = \left(\sum_{i=1}^n (y_i - \mathbf{w}^T\mathbf{x}_i)^2 \right) + \left(\alpha \sum_{j=1}^p w_j^2 \right)$ can be written in vector form as 

$\mathcal{L}(\mathbf{w})= \dfrac{1}{n}(\mathbf{y}-\mathbf{X}\mathbf{w})^T(\mathbf{y}-\mathbf{X}\mathbf{w})+\alpha \mathbf{w}^{T}\mathbf{w} $ , 

which is equal to the following equation with the $L_2$ norm,

$\mathcal{L}(\mathbf{w})=\dfrac{1}{n}\mathbf{w}^T\mathbf{X}^T\mathbf{X}\mathbf{w}-\dfrac{2}{n}\mathbf{w}^T\mathbf{X}^T\mathbf{y}+\dfrac{1}{n}\mathbf{y}^T\mathbf{y}+\alpha|w|_2$.

2- Taking partial derivatives with respect to $\mathbf{w}$ we get,


$\dfrac{\partial \mathcal{L}}{\partial \mathbf{w}}= \dfrac{2}{n}\mathbf{X}^T\mathbf{X}\mathbf{w}-\dfrac{2}{n}\mathbf{X}^T\mathbf{y}+2\alpha\mathbf{w}$. 

This was derived using the identity $\dfrac{\partial\mathbf{w}^T\mathbf{w}}{\partial \mathbf{w}} =2\mathbf{w}$ on the first term on the right hand side, since $\mathbf{w}^T\mathbf{X}^T\mathbf{X}\mathbf{w}=\mathbf{w}^T\mathbf{w}\mathbf{X}^T\mathbf{X}$. Then the identity $\dfrac{\partial\mathbf{w}^T\mathbf{X}^T}{\partial \mathbf{w}} =\mathbf{X}^T$ was used, which resulted in the second term on the right hand side. Lastly the third term was derived with the same identity as the first term.

3- Since $\mathcal{L}(\mathbf{w})$ is convex in $\mathbf{w}$, one can set the gradient equal to zero and solve for $\mathbf{w}$, i.e setting the gradient gradient of a convex function equal to zero solves for the minimum of the function. 


$\dfrac{\partial \mathcal{L}}{\partial \mathbf{w}}= \dfrac{2}{n}\mathbf{X}^T\mathbf{X}\mathbf{w}-\dfrac{2}{n}\mathbf{X}^T\mathbf{y}+2\alpha\mathbf{w}=0$.

This can be simplified as,

$\dfrac{2}{n}\mathbf{X}^T\mathbf{X}\mathbf{w}+2\alpha\mathbf{w}=\dfrac{2}{n}\mathbf{X}^T\mathbf{y}$,

which can rewritten as,

$(\mathbf{X}^T\mathbf{X}+n\alpha I)\mathbf{w}=\mathbf{X}^T\mathbf{y}$.

The identity matrix $I$ is included in order to get the right dimensions since $\alpha$ and $n$ are scalars. Then one can solve for $\mathbf{w^*}$, which results in the following expression,

$\mathbf{w^*}=(\mathbf{X}^T\mathbf{X}+n\alpha I)^{-1}\mathbf{X}^T\mathbf{y}$


4- The solution $\mathbf{w^*}$ is unique for $\alpha$ implying that $(\mathbf{X}^T\mathbf{X}+n\alpha I)$ is invertible. A matrix is invertible if it is positive definite. So, to find an $\alpha$ for which the solution is unique, one searches for an $\alpha$ making the matrix positive definite. 

For any $\mathbf{v}\in R^d$,  

$\mathbf{v}^T(\mathbf{X}\mathbf{X}^T)\mathbf{v}=(\mathbf{X}^T\mathbf{v})^T(\mathbf{X}^T\mathbf{v})\geq 0$ 

and $n\alpha I$ is already positive definite since $n,\alpha > 0$. 

Thus it follows that, 

$\mathbf{v}^T(\mathbf{X}\mathbf{X}^T+n\alpha I)\mathbf{v}=\mathbf{v}^T(\mathbf{X}\mathbf{X}^T)\mathbf{v}+\mathbf{v}^T n\alpha I\mathbf{v}\geq \mathbf{v}^T(n\alpha I)\mathbf{v} > 0$ 

which for $n,\alpha>0$ makes $(\mathbf{X}\mathbf{X}^T+n\alpha I)$ positive definite and invertible and the solution $\mathbf{w^*}$ unique. 



```python
# Make it possible to show plots in the notebooks.
%matplotlib inline
import numpy as np
from sklearn import datasets
from sklearn.preprocessing import normalize
import matplotlib
import matplotlib.pyplot as plt



def fit_linear_with_regularization(X, y, alpha):
    """ This function takes as input the features, the targets and the regularisation parameter
    alpha and performs linear regression with regularisation. 
    The function returns the least squares solution 
    """
    
    w=(np.linalg.inv(((np.transpose(X)).dot(X)+X.shape[0]*alpha*np.identity(X.shape[1])))).dot(np.transpose(X).dot(y))
    
    return w 

def predict(X_test, w):
    """ This functino takes as input a dataset  (in the same dimensions as ) 
    and returns the predictions.
    """
    y_pred=X_test.dot(w)
    
    return y_pred


def plot_prediction(X_test, y_test, y_pred):
    """Function that scatter plots of the first feature of  and the diabetes progression ytest
    and plots the prediction. Then it calculates the mean squared error that explains the accuracy of the predictions
    """
  # Scatter plot the first feature of X_test (x-axis) and y_test (y-axis)
    
    matplotlib.pyplot.scatter(X_test[:,1],y_test,label='The first (non constant) feature of the test data')
  #plt.plot(X_test) # TODO change me
  
  # Plot y_pred using the first feature of X_test as x-axis
    matplotlib.pyplot.scatter(X_test[:,1],y_pred, label='The prediction using the first (non constant) feature of the test data')
    
    plt.title('Scatter plot of the first feature of Xtest and ytest as well as the prediction for Xtest')

  # Compute the mean squared error
    mean_squared_error = np.mean((y_test-y_pred)**2)

    return mean_squared_error


# Load the diabetes dataset
X, y = datasets.load_diabetes(return_X_y=True)

#normalize data
X=normalize(X)

#add constant feature "1" to the dataset
oness = np.ones((X.shape[0],1))
X = np.hstack((oness,X))

# Split the dataset into training and test set
num_test_elements = 20

X_train = X[:-num_test_elements]
X_test = X[-num_test_elements:]

y_train = y[:-num_test_elements]
y_test = y[-num_test_elements:]

start = 0.001
stop = 0.011
step = 0.001

float_range_array = np.arange(start, stop, step)
float_range_list = list(float_range_array)

# Set alpha
i = float_range_list

error_list=[]

# Train using linear regression with regularization and find optimal model
#loop over different values of alpha in order to find the one resulting in the lowest mean squared error
for alpha in i:
    w = fit_linear_with_regularization(X_train, y_train, alpha)
    # Make predictions using the testing set X_test
    y_pred = predict(X_test, w)
    # Plots and mean squared error
    error = plot_prediction(X_test, y_test, y_pred)
    # Show the plot
    plt.legend()
    plt.show();
    print("For alpha = {}, w is {})".format(alpha, w))
    print('The mean squared error is ', error)
    
    #append error to list
    error_list.append(error)
    
print('The total smallest mean squared error is ', min(error_list),'for alpha=0.008')


```

7- The Gaussian prior is defined as $p(\mathbf{w}) = \mathcal{N}(\mathbf{0}, \lambda^{-1} \mathbf{I})$ and the likelihood as $p(\mathbf{y} | \mathbf{X}, \mathbf{w}) = \prod_{i=1}^n \mathcal{N}(\mathbf{w}^T \mathbf{x}_i, \gamma^{-1})$. To find the posterior we can multiply the prior with the likelihood since the normalizing constants can be ignored. This results in the following equation,

$p(\mathbf{w} | \mathbf{y}, \mathbf{X})=p(\mathbf{w})p(\mathbf{y} | \mathbf{X}, \mathbf{w}) \propto \exp(-\dfrac{\gamma}{2}(y-Xw)^T(y-Xw))\exp(-\dfrac{\lambda}{2}w^Tw)$.

Then, taking the logarithm results in the following equation for the log posterior,

$\ln(p(\mathbf{w} | \mathbf{y}, \mathbf{X})=-\dfrac{\gamma}{2}(y-Xw)^T(y-Xw)-\dfrac{\lambda}{2}w^Tw$


8- The function for $\mathcal{L}(\mathbf{w})$ is given by

$\mathcal{L}(\mathbf{w})= \dfrac{1}{n}(\mathbf{y}-\mathbf{X}\mathbf{w})^T(\mathbf{y}-\mathbf{X}\mathbf{w})+\alpha \mathbf{w}^{T}\mathbf{w} $.

From the above expression for the log posterior, one can see that their relation is 

$\ln(p(\mathbf{w} | \mathbf{y}, \mathbf{X}) \propto -\mathcal{L}(\mathbf{w})$.

This is obtained by setting $\alpha = \dfrac{\lambda}{\gamma}$ in the equation for $\mathcal{L}(\mathbf{w})$ and simplifying the expression for $\ln(p(\mathbf{w} | \mathbf{y}, \mathbf{X})$ by multiplying with 2 and dividing by $\gamma$. Since the purpose is to maximize the log posterior, it is set equal to 0 and thereby these operations are allowed. 

This results in the following expressions,

$-\mathcal{L}(\mathbf{w})\propto-\dfrac{1}{n}(\mathbf{y}-\mathbf{X}\mathbf{w})^T(\mathbf{y}-\mathbf{X}\mathbf{w})-\dfrac{\lambda}{\gamma} \mathbf{w}^{T}\mathbf{w}$

$\ln(p(\mathbf{w} | \mathbf{y}, \mathbf{X})\propto-(y-Xw)^T(y-Xw)-\dfrac{\lambda}{\gamma}w^Tw$
