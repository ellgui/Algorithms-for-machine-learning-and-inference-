## Setting hyperparameters
A beta distribution, $Beta(\alpha,\beta)$, with shape parameters $\alpha$ and $\beta$ has the expected value given by $$E[\theta]= \frac{\alpha}{\alpha+\beta}$$ and the variance given by $$Var[\theta]=\frac{\alpha \beta}{(\alpha+\beta)^{2}(\alpha + \beta +1)}.$$


Given that $\theta \sim Beta(a^{2},b)$ it follows that the shape parameters $\alpha$ and $\beta$ equals $\alpha=a^{2}$ and $\beta=b$. Inserting $\alpha=a^{2}$ and $\beta=b$, and the given expected value and variance one obtains $$E[\theta]=\frac{a^{2}}{a^{2}+b}=m$$ and $$Var[\theta]=\frac{a^{2} b}{(a^{2}+b)^{2}(a +b +1)}=\nu.$$ From the equation for the expected value it follows that $$b=a^{2}(\frac{1}{m}-1).$$ In the same way, the expression for $a$ can be obtained from the expression for the variance with the obtained expression for b inserted as  $$\frac{a^{2}(a^{2}(\frac{1}{m}-1))}{(a^{2}+(a^{2}(\frac{1}{m}-1))^{2}(a^{2}+a^{2}(\frac{1}{m}-1)+1)}=\nu,$$ which one can simplify and obtain $$a=\sqrt{m(\frac{m-m^{2}}{v}-1)}.$$ Inserting the obtained expression for $a$ in the expression for $b$ one obtains $$b=m(\frac{m-m^{2}}{\nu}-1)(\frac{1}{m}-1)$$


## Correlation
Since $X$ is a continuous random variable, uniformly distributed in $[-2, +2]$, the formula for the expected value is,

$E[X]=\int_a^b xf(x)dx$.

Since $X$ is uniformly distributed in $[-2, +2]$, the probability distribution function $f(x)$ can be written as,

$f(x)=\frac{1}{b-a}=\frac{1}{2-(-2)}=\frac{1}{4}.$

The expected value can then be calculated as,

$E[X]=\int_2^{-2} x\frac{1}{4}dx=0$, for a continous random variable that is uniformly distributed.

The formula for the covariance is,

$cov(X,Y)=E[(X-E[X])(Y-E[Y])]$,

and can be rewritten as,

$cov(X,Y) = E[(X-E[X])(Y-E[Y])] = E[XY-XE[Y]-E[X]Y+E[X]E[Y]]$.

Since $E[E[X]]=E[X]$ expression can bli simplified as,

$E[XY-XE[Y]-E[X]Y+E[X]E[Y]] = E[XY] - E[X]E[Y]- E[X]E[Y]+E[X]E[Y] = E[XY]-E[X]E[Y]$.

We also know that $Y=X^4$, which can simplify the equation further as,

$E[XY]-E[X]E[Y] = E[X\cdot X^4]-E[X]E[X^4] = E[X^5]-E[X]E[X^4]$,

where $E[X^5]=\int_2^{-2} x^5\frac{1}{4}dx=0$ and $E[X]=0$, resulting in the final expression,

$cov(X,X^4)= E[X^5]-E[X]E[X^4]=0-0\cdot E[X^4]=0$

We have $ \mathcal{N}(\mathbf{X} | \mu, \Sigma)$, where the covariance matrix $\Sigma=\operatorname{diag}\left(\sigma_{1}^{2}, \sigma_{2}^{2}, \ldots, \sigma_{n}^{2}\right)$.

In our case $N(\mu, 2\sigma^2I)$, i.e $\Sigma=2\sigma^2I$.

In order to find the maximum likelihood, one needs to find $\mu$ and $\Sigma$ that maximises the following expression,

$L(X | \mu, \Sigma)=\prod_{n=1}^{N}p(x_n ; \mu, \Sigma)$.

The multivariate Gaussian distribution becomes,

$p\left(\mathbf{x}_{n} | \sigma^{2}, \mu\right)=\Pi_{n=1}^{N}\left(\frac{1}{(2 \pi)^{1 / 2} \mid 2 \sigma^{2} \mid^{1/2}}\right) \exp \left(-\frac{1}{4 \sigma^{2}}\left(\mathbf{x}_{n}-\mu\right)^{T}\left(\mathbf{x}_{n}-\mu\right)\right)$.

To simplify the calculation, it is possible to take the logarithm of the likelihood function. 

$\log(L)=\sum_{n=1}^{N}\left(-\frac{1}{2} \log (2 \pi)-\frac{1}{2} \log \left(\mid2\sigma^{2} \mid\right)-\frac{1}{4 \sigma^{2}}\left(\mathbf{x}_{n}-\mu\right)^{T}\left(\mathbf{x}_{n}-\mu\right)\right)$.

To find the maximum of the log likelihood, one takes the partial derivative of the equation with respect to $\mu$ and with respect to $\sigma$ respectively. By setting the partial derivates equal to 0 and using the partial derivative rule $\frac{\partial W^{T} A W}{\partial W}=2 A W$, the optimal value of each parameter is found, which is presented below,


$\frac{\partial \log (L)}{\partial \boldsymbol{\mu}}=\frac{1}{4 \sigma^{2}} \sum_{n=1} 2\left(\mathbf{x}_{n}-\hat{\boldsymbol{\mu}}\right)=N \hat{\boldsymbol{\mu}}-\sum_{n=1}^{N} \mathbf{x}_{n}=0$,

resulting in the following expresssion,


$\begin{aligned} \hat{\mu} &=\frac{1}{N} \sum_{n=1}^{N} \mathbf{x}_{n}\end{aligned}$. 

Similar calculations were done for $\sigma$,

$\frac{\partial \log (L)}{\partial \sigma}=-\frac{N}{\hat{\sigma}}+\frac{1}{2 \hat\sigma^{3}} \sum_{n=1}^{N}\left(\mathbf{x}_{n}-\boldsymbol{\mu}\right)^{T}\left(\mathbf{x}_{n}-\boldsymbol{\mu}\right)=0$,

and

$\hat{\sigma}^{2}=\frac{1}{2 N} \sum_{n=1}^{N}\left(\mathbf{x}_{n}-\boldsymbol{\mu}\right)^{T}\left(\mathbf{x}_{n}-\boldsymbol{\mu}\right)$, 

resulting in, 

$\hat{\sigma}=\sqrt{\frac{1}{2 N} \sum_{n=1}^{N}\left(\mathbf{x}_{n}-\boldsymbol{\mu}\right)^{T}\left(\mathbf{x}_{n}-\boldsymbol{\mu}\right)}$.

```python
# Sample Answer to (b)
import numpy as np # numpy


def sge(X):
    """
    This function computes the mean mu and standard deviation sigma 
    for a dataset coming from spherical Gaussian distribution N(mu, 2 sigma^2 I)

    @param X : A numpy array of size n x p where each row represents a p-dimensional data point
    e.g. X = [ [2, 1], [3, 7], [4 ,5] ] is a dataset with 3 samples having two coordinates each.
    @return mu: A numpy array of size p representing the estimated mean of the dataset
    @return sigma: A real number indicating the estimated standard deviation of the dataset
    """
    n=X.shape[0]
    p=X.shape[1]
    mu = 0
    
    for x_n in X:
        mu += x_n / n      
    D = X - mu
    
    sigmaSquare = 0
    for diff in D:
        sigmaSquare += diff.dot(np.transpose(diff))
        
    sigma = np.sqrt(sigmaSquare/(2*n*p))
    
    return mu, sigma
```
